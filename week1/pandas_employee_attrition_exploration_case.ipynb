## 7. Create missing values for teaching practice

**Task:** The original dataset has no major missing-value problem. To practice missing-value handling, create a teaching copy with artificial missing values.

**Goal:** Practice realistic cleaning strategies:

- fill numerical missing values with group median
- fill categorical missing values with a placeholder such as `Unknown`
- verify that missing values were handled correctly

This section is for teaching only. In a real analysis, you should not randomly create missing values.

```python
practice_df = df_work.copy()

# Use a fixed random seed so everyone gets the same missing values.
rng = np.random.default_rng(seed=42)

# Randomly set about 3% of MonthlyIncome values to missing.
income_missing_index = rng.choice(practice_df.index, size=int(0.03 * len(practice_df)), replace=False)
practice_df.loc[income_missing_index, "MonthlyIncome"] = np.nan

# Randomly set about 2% of JobSatisfaction values to missing.
satisfaction_missing_index = rng.choice(practice_df.index, size=int(0.02 * len(practice_df)), replace=False)
practice_df.loc[satisfaction_missing_index, "JobSatisfaction"] = np.nan

# Randomly set about 1% of BusinessTravel values to missing.
travel_missing_index = rng.choice(practice_df.index, size=int(0.01 * len(practice_df)), replace=False)
practice_df.loc[travel_missing_index, "BusinessTravel"] = np.nan

practice_df[["Department", "MonthlyIncome", "JobSatisfaction", "BusinessTravel"]].isna().sum()
```

## 8. Handle missing values

**Task:** Fill the artificial missing values using reasonable rules.

**Business logic:**

- Monthly income varies by department, so fill missing income using the **department median**.
- Job satisfaction is an ordinal score, so fill missing values using the overall median.
- Business travel is categorical, so fill missing values with `Unknown`.

**Pandas concepts:** `groupby()`, `transform()`, `lambda`, `fillna()`.

```python
# Before cleaning, inspect rows where MonthlyIncome is missing.
practice_df.loc[practice_df["MonthlyIncome"].isna(), ["Department", "JobRole", "MonthlyIncome"]].head()
```

```python
# Fill MonthlyIncome with the median MonthlyIncome within each Department.
practice_df["MonthlyIncome"] = practice_df.groupby("Department")["MonthlyIncome"].transform(
    lambda income_series: income_series.fillna(income_series.median())
)

# Fill JobSatisfaction with the overall median.
practice_df["JobSatisfaction"] = practice_df["JobSatisfaction"].fillna(practice_df["JobSatisfaction"].median())

# Fill missing BusinessTravel values with Unknown.
practice_df["BusinessTravel"] = practice_df["BusinessTravel"].fillna("Unknown")

# Verify remaining missing values.
practice_df[["MonthlyIncome", "JobSatisfaction", "BusinessTravel"]].isna().sum()
```

```python
# Continue the analysis with the cleaned practice dataframe.
df_clean = practice_df.copy()

print("Cleaned dataframe shape:", df_clean.shape)
df_clean.head()
```

## 11. Create new columns for easier analysis

**Business question:** Can we convert raw columns into more useful business categories?

**Task:** Create several new columns:

1. `IncomeK`: monthly income in thousands
2. `OvertimeFlag`: numerical version of overtime
3. `YoungEmployee`: whether age is below 30
4. `LongTenure`: whether employee has worked at the company for at least 10 years
5. `IncomePerYearAtCompany`: income divided by years at company, with special handling for zero years

**Pandas concepts:** column arithmetic, `np.where()`, boolean conditions.

```python
df_clean["IncomeK"] = df_clean["MonthlyIncome"] / 1000

df_clean["OvertimeFlag"] = df_clean["OverTime"].map({"Yes": 1, "No": 0})

df_clean["YoungEmployee"] = np.where(df_clean["Age"] < 30, "Yes", "No")

df_clean["LongTenure"] = np.where(df_clean["YearsAtCompany"] >= 10, "Yes", "No")

# Avoid division by zero when YearsAtCompany is 0.
df_clean["IncomePerYearAtCompany"] = np.where(
    df_clean["YearsAtCompany"] == 0,
    np.nan,
    df_clean["MonthlyIncome"] / df_clean["YearsAtCompany"]
)

df_clean[["Age", "MonthlyIncome", "IncomeK", "OverTime", "OvertimeFlag", "YoungEmployee", "LongTenure", "IncomePerYearAtCompany"]].head(10)
```

## 12. Create categories using a custom function and `apply()`

**Business question:** Can we create a simple employee life-stage category based on age?

**Task:** Write a function that assigns employees into age bands.

**Pandas concepts:** custom function, `apply()`.

```python
def assign_age_group(age):
    """Return an age group label based on employee age."""
    if age < 30:
        return "Under 30"
    elif age < 40:
        return "30-39"
    elif age < 50:
        return "40-49"
    else:
        return "50+"

# Apply the function to the Age column.
df_clean["AgeGroup"] = df_clean["Age"].apply(assign_age_group)

df_clean[["Age", "AgeGroup"]].head(10)
```

## 13. Create a risk segment using row-wise `apply()`

**Business question:** Which employees might deserve closer HR attention based on simple business rules?

**Task:** Create a rule-based `RetentionRiskSegment` using multiple columns.

Example business rules:

- High Risk: employee works overtime and has low job satisfaction
- High Risk: employee is young, works overtime, and has low income
- Medium Risk: employee has low environment satisfaction or low work-life balance
- Low Risk: otherwise

This is not a predictive model. It is a transparent business-rule exercise for pandas practice.

**Pandas concepts:** row-wise `apply(axis=1)`, custom function, multiple conditions.

```python
def assign_retention_risk(row):
    """Assign a simple rule-based retention risk segment."""
    if row["OverTime"] == "Yes" and row["JobSatisfaction"] <= 2:
        return "High Risk"
    elif row["Age"] < 30 and row["OverTime"] == "Yes" and row["MonthlyIncome"] < 4000:
        return "High Risk"
    elif row["EnvironmentSatisfaction"] <= 2 or row["WorkLifeBalance"] <= 2:
        return "Medium Risk"
    else:
        return "Low Risk"

# axis=1 means apply the function row by row.
df_clean["RetentionRiskSegment"] = df_clean.apply(assign_retention_risk, axis=1)

df_clean[[
    "Age", "OverTime", "MonthlyIncome", "JobSatisfaction", 
    "EnvironmentSatisfaction", "WorkLifeBalance", "RetentionRiskSegment", "Attrition"
]].head(15)
```

## 14. Use `lambda` for quick transformations

**Task:** Create compact labels using `lambda` functions.

**Goal:** Show when `lambda` is useful for short, simple transformations.

**Pandas concepts:** `apply(lambda x: ...)`.

```python
# Convert distance to a simple near/far label.
df_clean["DistanceCategory"] = df_clean["DistanceFromHome"].apply(
    lambda x: "Far" if x >= 10 else "Near"
)

# Create a short department code using string slicing.
df_clean["DepartmentCode"] = df_clean["Department"].apply(lambda x: x[:3].upper())

# Create a simple income level label.
df_clean["IncomeLevel"] = df_clean["MonthlyIncome"].apply(
    lambda x: "High" if x >= 10000 else ("Medium" if x >= 5000 else "Low")
)

df_clean[["Department", "DepartmentCode", "DistanceFromHome", "DistanceCategory", "MonthlyIncome", "IncomeLevel"]].head(10)
```

## 15. Explore attrition by one categorical variable

**Business question:** Which departments have higher attrition rates?

**Task:** Group employees by department and calculate:

- number of employees
- number of employees who left
- attrition rate
- average monthly income
- average job satisfaction

**Pandas concepts:** `groupby()`, `agg()`, named aggregation, sorting.

```python
department_summary = (
    df_clean
    .groupby("Department")
    .agg(
        employee_count=("AttritionFlag", "size"),
        attrition_count=("AttritionFlag", "sum"),
        attrition_rate=("AttritionFlag", "mean"),
        avg_monthly_income=("MonthlyIncome", "mean"),
        median_monthly_income=("MonthlyIncome", "median"),
        avg_job_satisfaction=("JobSatisfaction", "mean")
    )
    .sort_values("attrition_rate", ascending=False)
)

# Convert attrition rate to percentage for easier interpretation.
department_summary["attrition_rate"] = department_summary["attrition_rate"] * 100

department_summary.round(2)
```

## 16. Explore attrition by multiple categorical variables

**Business question:** Does overtime appear to matter across departments?

**Task:** Group by both `Department` and `OverTime`.

**Goal:** Compare attrition rates for overtime vs. non-overtime employees within each department.

**Pandas concepts:** multi-column `groupby()`, `agg()`, `reset_index()`.

```python
department_overtime_summary = (
    df_clean
    .groupby(["Department", "OverTime"])
    .agg(
        employee_count=("AttritionFlag", "size"),
        attrition_count=("AttritionFlag", "sum"),
        attrition_rate=("AttritionFlag", "mean"),
        avg_income=("MonthlyIncome", "mean"),
        avg_years_at_company=("YearsAtCompany", "mean")
    )
    .reset_index()
)

department_overtime_summary["attrition_rate"] = department_overtime_summary["attrition_rate"] * 100

department_overtime_summary.sort_values(["Department", "attrition_rate"], ascending=[True, False]).round(2)
```

## 17. Use `transform()` to compare each employee with their group

**Business question:** Is an employee paid above or below the average for their job role?

**Task:** For each employee, calculate:

- average monthly income for their job role
- difference between their income and job-role average
- whether they are below their job-role average

**Pandas concepts:** `groupby().transform()`, creating comparison columns.

```python
# transform returns a Series with the same length as the original dataframe.
df_clean["JobRoleAvgIncome"] = df_clean.groupby("JobRole")["MonthlyIncome"].transform("mean")

df_clean["IncomeVsJobRoleAvg"] = df_clean["MonthlyIncome"] - df_clean["JobRoleAvgIncome"]

df_clean["BelowJobRoleAvgIncome"] = np.where(df_clean["IncomeVsJobRoleAvg"] < 0, "Yes", "No")

df_clean[[
    "JobRole", "MonthlyIncome", "JobRoleAvgIncome", 
    "IncomeVsJobRoleAvg", "BelowJobRoleAvgIncome", "Attrition"
]].head(10)
```

```python
# Question: Do employees below their job-role average income have higher attrition?
income_gap_summary = (
    df_clean
    .groupby("BelowJobRoleAvgIncome")
    .agg(
        employee_count=("AttritionFlag", "size"),
        attrition_count=("AttritionFlag", "sum"),
        attrition_rate=("AttritionFlag", "mean"),
        avg_income_gap=("IncomeVsJobRoleAvg", "mean")
    )
)

income_gap_summary["attrition_rate"] = income_gap_summary["attrition_rate"] * 100
income_gap_summary.round(2)
```

## 18. Build a reusable summary function

**Task:** Write a function that summarizes attrition by any categorical column.

**Goal:** Avoid repeating the same groupby code again and again.

**Pandas concepts:** defining functions, `groupby()`, `agg()`, sorting, reuse.

```python
def summarize_attrition_by(category_col):
    """
    Summarize employee count, attrition count, and attrition rate by a categorical column.
    
    Parameters
    ----------
    category_col : str
        Column name used for grouping.
    
    Returns
    -------
    pandas.DataFrame
        Summary table sorted by attrition rate.
    """
    summary = (
        df_clean
        .groupby(category_col)
        .agg(
            employee_count=("AttritionFlag", "size"),
            attrition_count=("AttritionFlag", "sum"),
            attrition_rate=("AttritionFlag", "mean"),
            avg_income=("MonthlyIncome", "mean"),
            avg_age=("Age", "mean")
        )
        .sort_values("attrition_rate", ascending=False)
    )
    
    summary["attrition_rate"] = summary["attrition_rate"] * 100
    return summary.round(2)
```

```python
# Example 1: attrition by job role
summarize_attrition_by("JobRole")
```

```python
# Example 2: attrition by age group
summarize_attrition_by("AgeGroup")
```

```python
# Example 3: attrition by income level
summarize_attrition_by("IncomeLevel")
```

## 20. Find the highest-risk groups using grouped filters

**Business question:** Which groups are large enough to matter and have high attrition rates?

**Task:** Summarize attrition by job role, then filter to groups with at least 30 employees.

**Goal:** Avoid overreacting to tiny groups.

**Pandas concepts:** `groupby()`, `agg()`, filtering summary tables with `loc`.

```python
jobrole_summary = summarize_attrition_by("JobRole")

# Filter to job roles with at least 30 employees.
large_jobrole_summary = jobrole_summary.loc[jobrole_summary["employee_count"] >= 30]

large_jobrole_summary.sort_values("attrition_rate", ascending=False)
```

## 21. Create a pivot-style comparison table without visualization

**Business question:** How does attrition rate vary by age group and overtime status?

**Task:** Create a table where rows are age groups, columns are overtime status, and values are attrition rates.

## 22. More complex case: build an HR action table

**Business question:** Which employee segments may deserve immediate HR review?

**Task:** Build a table by `RetentionRiskSegment`, `OverTime`, and `IncomeLevel`.

For each segment, calculate:

- number of employees
- attrition count
- attrition rate
- average monthly income
- average job satisfaction
- average years at company

Then keep only groups with at least 10 employees and sort by attrition rate.

**Pandas concepts:** multi-column groupby, named aggregation, filtering, sorting.

```python
hr_action_table = (
    df_clean
    .groupby(["RetentionRiskSegment", "OverTime", "IncomeLevel"])
    .agg(
        employee_count=("AttritionFlag", "size"),
        attrition_count=("AttritionFlag", "sum"),
        attrition_rate=("AttritionFlag", "mean"),
        avg_income=("MonthlyIncome", "mean"),
        avg_job_satisfaction=("JobSatisfaction", "mean"),
        avg_years_at_company=("YearsAtCompany", "mean")
    )
    .reset_index()
)

hr_action_table["attrition_rate"] = hr_action_table["attrition_rate"] * 100

hr_action_table = hr_action_table.loc[hr_action_table["employee_count"] >= 10]

hr_action_table.sort_values("attrition_rate", ascending=False).round(2)
```

## 23. Identify individual records for follow-up

**Business question:** Which employees match a high-risk profile?

**Task:** Use `loc` to find employees who:

- are in the High Risk segment
- work overtime
- have low or medium income
- have low job satisfaction

**Pandas concepts:** compound filtering, `loc`, selecting useful columns, sorting.

```python
high_risk_employees = df_clean.loc[
    (df_clean["RetentionRiskSegment"] == "High Risk") &
    (df_clean["OverTime"] == "Yes") &
    (df_clean["IncomeLevel"].isin(["Low", "Medium"])) &
    (df_clean["JobSatisfaction"] <= 2),
    [
        "Age", "Department", "JobRole", "MonthlyIncome", "IncomeLevel",
        "OverTime", "JobSatisfaction", "EnvironmentSatisfaction", 
        "WorkLifeBalance", "YearsAtCompany", "Attrition"
    ]
]

high_risk_employees.sort_values(["Attrition", "MonthlyIncome"], ascending=[False, True]).head(20)
```

## 24. Use `assign()` and method chaining

**Task:** Rewrite part of the analysis using method chaining.

**Goal:** Show a more advanced pandas style where multiple steps are connected.

This style is common in professional pandas code, but students should first understand the step-by-step version.

```python
chained_summary = (
    df_clean
    .assign(
        IncomeBelow5000=lambda data: data["MonthlyIncome"] < 5000,
        ShortTenure=lambda data: data["YearsAtCompany"] < 3
    )
    .groupby(["IncomeBelow5000", "ShortTenure", "OverTime"])
    .agg(
        employee_count=("AttritionFlag", "size"),
        attrition_count=("AttritionFlag", "sum"),
        attrition_rate=("AttritionFlag", "mean"),
        avg_age=("Age", "mean")
    )
    .reset_index()
    .sort_values("attrition_rate", ascending=False)
)

chained_summary["attrition_rate"] = chained_summary["attrition_rate"] * 100
chained_summary.round(2)
```
