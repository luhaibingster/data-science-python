# Pandas Part 1: Exploring Real Data with Pandas

In this session, students will not just learn Pandas syntax. They will learn how to **approach a real analytics problem** using a dataset that looks and feels like real business data.

### Business scenario

You are part of the analytics team for a growing specialty beverage retailer. Leadership wants answers to questions such as:

- Which countries and cities generate the most revenue?
- Which product categories sell the most?
- Which channels appear strongest?
- Are there data quality issues we should notice before deeper analysis?
- What should management investigate next?

This session focuses on the very beginning of the analytics workflow:

> **load → inspect → understand → question → explore**

---

## Learning objectives

By the end of this session, students should be able to:

- Load a dataset into a Pandas DataFrame
- Inspect its basic structure, columns, data types, and size
- Distinguish between rows, columns, index, and values
- Use `head()`, `tail()`, `sample()`, `info()`, `describe()`, and `value_counts()`
- Select columns and filter rows
- Sort and summarize data to answer practical business questions
- Identify potential data quality issues before cleaning
- Document observations in a notebook clearly and professionally

## Core inspection commands

These are among the most important commands in all of Pandas:

- `head()` → first few rows

Students should become fluent with these early.

### Why `info()` matters

`info()` is often the first serious inspection tool an analyst uses because it quickly shows:
- total rows
- total columns
- data types
- missing values
- approximate memory usage

## Frequency tables for categorical columns

In practical analytics work, a very fast way to understand a dataset is to ask:

- How many countries?
- Which channels appear most often?
- Which categories are most common?
- Which city values look suspicious?

The workhorse function here is:

```python
value_counts()
```


````python
df["country"].value_counts()
````

````python
df["category"].value_counts(dropna=False) # NaN values are included in the count
````

````python
df["channel"].value_counts(dropna=False, ascending=True) 
````

````python
df["city"].value_counts(dropna=False).head(15)
````

### Teaching moment

Notice that `value_counts(dropna=False)` includes missing values.  
This is extremely useful because missingness is itself information.

### Discussion prompts

## Business question: Which categories appear most often?

This is not the same as highest revenue.  
Frequency and value are different.

A category might appear often but generate lower revenue per transaction.


````python
df["category"].value_counts(dropna=False)
````

````python
df.groupby("category", dropna=False)["revenue"].sum().sort_values(ascending=False)
````

### Question

Compare:
- category frequency
- category revenue

Do the rankings match? If not, what might explain the difference?


## Business question: Which channels seem strongest?

Channels matter for operational strategy, cost structure, and customer experience.


````python
df["channel"].value_counts(dropna=False)
````

````python
df.groupby("channel", dropna=False)["revenue"].agg(["count", "sum", "mean"]).sort_values("sum", ascending=False)
````

### Question

What is the difference between:
- number of transactions (`count`)
- total revenue (`sum`)
- average revenue per transaction (`mean`)

## A closer look at dates

Dates are crucial in real analytics work.

Even in an exploration session, students should check:
- Is the date column stored as text or actual datetime?
- What is the date range?
- Are there obvious gaps or impossible values?


````python
df["order_date"].head()
````

````python
df["order_date"] = pd.to_datetime(df["order_date"])
df["order_date"].dtype
````

````python
df["order_date"].min(), df["order_date"].max()
````

````python
df["order_month"] = df["order_date"].dt.to_period("M")
df[["order_date", "order_month"]].head()
````

## Customer-level exploration

A common analytics question is whether revenue comes from many small customers or a few large ones.


````python
customer_summary = (
    df.groupby("customer_id")
      .agg(
          transactions=("order_id", "count"),
          total_revenue=("revenue", "sum"),
          avg_revenue=("revenue", "mean")
      )
      .sort_values("total_revenue", ascending=False)
)

customer_summary.head(10)
````

### Question

What would leadership want to know next about top customers?
- country?
- segment?
- product preferences?
- retention over time?

## Solution sketch for guided lab

Use or reveal this section only after students attempt the guided lab.


````python
print("Rows, Columns:", df.shape)
print("\nMissing values:")
print(df.isna().sum())
print("\nNumeric columns:")
print(df.select_dtypes(include=np.number).columns.tolist())
print("\nCategorical/object columns:")
print(df.select_dtypes(exclude=np.number).columns.tolist())
````

````python
print("Country with most transactions:")
print(df["country"].value_counts().head(1))

print("\nCategory with highest total revenue:")
print(df.groupby("category")["revenue"].sum().sort_values(ascending=False).head(1))

print("\nChannel with highest average revenue:")
print(df.groupby("channel")["revenue"].mean().sort_values(ascending=False).head(1))

print("\nMost common city:")
print(df["city"].value_counts(dropna=False).head(5))
````

````python
print("Duplicate rows:", df.duplicated().sum())
print("Negative quantities:", (df["quantity"] < 0).sum())
print("\nUnique category values:")
print(df["category"].unique())
print("\nBlank-like city rows:")
print((df["city"].astype(str).str.strip() == "").sum())
````

## More challenging questions

- Find the top 10 products by total revenue.

- Compare average discount by category.
- Create a table showing transaction count, total revenue, and average revenue by:
    - country

````python
# Q1
df.groupby("product_name")["revenue"].sum().sort_values(ascending=False).head(10)
````

````python
# Q2
top_customers_by_country = (
    df.groupby(["country", "customer_id"])["revenue"]
      .sum()
      .reset_index()
      .sort_values(["country", "revenue"], ascending=[True, False])
)

top_customers_by_country.groupby("country").head(5)
````

````python
# Q3
df.groupby("category")["discount"].mean().sort_values(ascending=False)
````

````python
# Q4
df.groupby(["country", "channel", "category"])["revenue"].agg(["count", "sum", "mean"]).head(20)
````

````python
# Q5
df.loc[df["quantity"] < 0, ["order_id", "quantity", "revenue"]].head(20)
````

````python
# Q6
df[df["city"].astype(str).str.strip() == ""][["order_id", "country", "city", "revenue"]].head()
````
