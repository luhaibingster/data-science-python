# Data Visualization Tutorial (Beginner-Friendly)

This notebook introduces matplotlib and seaborn from basics to simple examples.

## 1. Import Libraries

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## 2. Matplotlib Basics

```
# Simple line plot
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```
# Simple bar chart
categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)
plt.title('Bar Chart')
plt.show()
```

## 3. Seaborn Basics

### Create a simple dataset

```
df = pd.DataFrame({
    'category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'sales': [10, 12, 8, 9, 14, 15],
    'region': ['East', 'West', 'East', 'West', 'East', 'West']
})

df
```

### Basic barplot

```
sns.barplot(data=df, x='category', y='sales')
plt.show()
```

### Barplot with hue

```
sns.barplot(data=df, x='category', y='sales', hue='region')
plt.show()
```

### Boxplot

```
sns.boxplot(data=df, x='category', y='sales')
plt.show()
```

### Scatterplot

```
sns.scatterplot(data=df, x='category', y='sales', hue='region')
plt.show()
```
