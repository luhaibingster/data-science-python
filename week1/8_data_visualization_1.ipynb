## Matplotlib vs Seaborn

We will use two main libraries:

### Matplotlib
Matplotlib is the **foundation** plotting library in Python.

It gives you:
- direct control over plots
- basic chart types such as line, bar, scatter, and histogram
- flexibility for labels, titles, axes, and figure size

Think of Matplotlib as the **base tool**.

### Seaborn
Seaborn is built **on top of Matplotlib**.

It gives you:
- cleaner syntax for many common charts
- better default appearance
- strong support for **pandas DataFrames**
- built-in handling for grouped statistical charts

Think of Seaborn as the **higher-level tool**.

### Simple comparison

| Library | Best for |
|---|---|
| Matplotlib | learning plotting basics, fine control, custom formatting |
| Seaborn | faster statistical charts, grouped comparisons, cleaner defaults |

### Practical advice
- Learn **Matplotlib first** so the core ideas make sense.
- Then learn **Seaborn** because it is very convenient for real analytics work.

## Start with imports

We use:
- `pandas` for data tables
- `numpy` for simple numerical arrays
- `matplotlib.pyplot` for plotting
- `seaborn` for higher-level plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 50)
sns.set_theme()

print("Libraries imported successfully.")
```

## The most basic Matplotlib line chart

Start simple.

The core idea is:
- give the x values
- give the y values
- call `plt.plot(x, y)`
- then show the result

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 6, 7]

plt.plot(x, y)
plt.show()
```

### Add a title and axis labels

Now we add a little more structure.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 6, 7]

plt.plot(x, y)
plt.title("Simple Line Chart")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

### Add a few more arguments 

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 6, 7]

plt.figure(figsize=(7, 4))
plt.plot(x, y, marker="o", linestyle="--")
plt.title("Line Chart with Marker and Figure Size")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

## Basic Matplotlib bar chart

Bar charts are useful when you want to compare groups.

```python
categories = ["A", "B", "C", "D"]
values = [10, 18, 7, 14]

plt.bar(categories, values)
plt.show()
```

### Improve the bar chart

```python
categories = ["A", "B", "C", "D"]
values = [10, 18, 7, 14]

plt.figure(figsize=(7, 4))
plt.bar(categories, values)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

## Basic Matplotlib scatter plot

Scatter plots are useful for relationships between two numeric variables.

```python
x = [1, 2, 3, 4, 5, 6]
y = [3, 5, 4, 7, 8, 9]

plt.scatter(x, y)
plt.show()
```

### Improve the scatter plot

```python
x = [1, 2, 3, 4, 5, 6]
y = [3, 5, 4, 7, 8, 9]

plt.figure(figsize=(7, 4))
plt.scatter(x, y)
plt.title("Advertising Spend vs Sales")
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## Basic Matplotlib histogram

Histograms help us understand the distribution of one numeric variable.

```python
scores = [55, 60, 61, 62, 65, 67, 68, 70, 72, 74, 75, 77, 80, 82, 84, 90]

plt.hist(scores)
plt.show()
```

### Improve the histogram

```python
scores = [55, 60, 61, 62, 65, 67, 68, 70, 72, 74, 75, 77, 80, 82, 84, 90]

plt.figure(figsize=(7, 4))
plt.hist(scores, bins=6)
plt.title("Distribution of Exam Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

## A quick summary of core Matplotlib chart types

These are the most common beginner charts:

- `plt.plot(x, y)` → line chart
- `plt.bar(x, y)` → bar chart
- `plt.scatter(x, y)` → scatter plot
- `plt.hist(x)` → histogram

Common helper methods:

- `plt.title()` → chart title
- `plt.xlabel()` / `plt.ylabel()` → axis labels
- `plt.figure(figsize=(w, h))` → figure size
- `plt.grid(True)` → grid lines
- `plt.show()` → display chart

```python
df = pd.DataFrame({
    'category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'sales': [10, 12, 8, 9, 14, 15],
    'profits': [2, 3, 2, 2, 4, 5],
    'region': ['East', 'West', 'East', 'West', 'East', 'West']
})

df
```

## A first Seaborn chart

Seaborn works especially well with DataFrames.

Basic pattern:

- `data=df`
- `x="column_name"`
- `y="column_name"`

This is an important syntax pattern in Seaborn.

```python
sns.barplot(data=df, x='category', y='sales')
plt.show()
```

### Barplot with hue

The `hue` argument adds one more grouping variable.

This is very common in Seaborn.

```python
sns.barplot(data=df, x='category', y='sales', hue='region')
plt.show()
```

### Boxplot

```python
sns.boxplot(data=df, x='category', y='sales')
plt.show()
```

### Scatterplot

```python
sns.scatterplot(data=df, x='sales', y='profits')
plt.show()
```

## Seaborn histogram

Seaborn also makes distributions easy to visualize.

```python
sns.histplot(data=df, x="sales")
plt.title("Distribution of Sales")
plt.show()
```

### Basic syntax patterns you should know

These patterns appear again and again in Seaborn:

- `data=...` → which DataFrame to use  
- `x=...` → column for x-axis  
- `y=...` → column for y-axis  
- `hue=...` → additional grouping variable  
- `plt.title(...)` → chart title  

This is why Seaborn is convenient for DataFrame-based work.

## Add arguments

Below is a progression for the same idea.

### Version 1: very basic

```python
sns.barplot(data=df, x="category", y="profits")
plt.show()
```

### Version 2: add title and labels

```python
sns.barplot(data=df, x="category", y="profits")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

### Version 3: add an extra grouping variable

```python
sns.barplot(data=df, x="category", y="profits", hue="region")
plt.title("Average Profit by Category and Segment")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

### Version 4: add more formatting

This is more advanced, but now it should feel understandable because we built up to it.

```python
plt.figure(figsize=(9, 5))
sns.barplot(data=df, x="category", y="profits", hue="region", estimator="mean", errorbar=None)
plt.title("Average Profit by Category and Segment")
plt.xlabel("Category")
plt.ylabel("Mean Profit")
plt.xticks(rotation=0)
plt.grid(axis="y", alpha=0.3)
plt.show()
```
