## 1. Motivation: Why NumPy?


### Problem:
You have 1 million data points. You need to scale them.

👉 Try Python vs NumPy


```python
x = [1, 2, 3]
print(x * 2)
```

```python

# Python loop
data = list(range(100000))
result = []
for x in data:
    result.append(x * 2)
```

```python

# NumPy
import numpy as np
data = np.arange(100000)
result = data * 2
```


### Reflection:
- Why is NumPy faster?
- What changes in how we think?


# 8. Case Study: Energy Load Analysis

You are given hourly energy load data.


```python

# Simulated energy data
load = np.random.normal(1000, 200, 1000)
```


### Tasks:
1. Compute average load
2. Find peak load hours
3. Detect anomalies (> mean + 2*std)


```python

mean = np.mean(load)
std = np.std(load)

peaks = load[load > mean]
anomalies = load[load > mean + 2*std]

mean, peaks[:5], anomalies[:5]
```
