## 1. Why NumPy?
Compare Python vs NumPy

```
# Python loop
prices = [100, 200, 300, 400]
new_prices = []
for p in prices:
    new_prices.append(p * 1.1)
new_prices
```

```
# NumPy vectorized
import numpy as np
prices = np.array([100, 200, 300, 400])
prices * 1.1
```

## 2. Creating Arrays

```
import numpy as np

a = np.array([1,2,3])
b = np.zeros((2,3))
c = np.arange(0,10,2)
d = np.linspace(0,1,5)

a, b, c, d
```

## 3. Vectorized Operations

```
a = np.array([1,2,3])
b = np.array([4,5,6])

a + b
a * 2
a ** 2
```

## 4. Indexing and Filtering

```
data = np.array([10, 60, 30, 80])
data[data > 50]
```

## 5. 2D Arrays

```
A = np.array([[1,2,3],[4,5,6]])
A[:,1]
```

## 6. Statistics

```
sales = np.array([100,200,300])
np.mean(sales), np.sum(sales), np.std(sales)
```

## 7. Broadcasting

```
salaries = np.array([5000,6000,7000])
salaries + 500
```

## 8. Performance Demo

```
import time
a = np.arange(1_000_000)

start = time.time()
b = a * 2
print("NumPy:", time.time() - start)
```

## 9. Practice

```
temps = np.array([70, 75, 80, 65, 90])

# 1. Convert to Celsius
celsius = (temps - 32) * 5/9

# 2. Filter > 30C
hot = celsius[celsius > 30]

# 3. Average
avg = np.mean(celsius)

celsius, hot, avg
```
